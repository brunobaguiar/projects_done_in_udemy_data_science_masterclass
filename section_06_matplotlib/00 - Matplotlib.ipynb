## Implementing the Figure Object

```python
# DATA
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
fig =plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title("Power of 4")

# SMALL AXES
axes2 = fig.add_axes([0.2,0.4,0.3,0.3])
axes2.plot(a,b)

axes2.set_xlim(4,5)
axes2.set_ylim(0,1000)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title("Zoomed In")

plt.show()
```

## Figure Parameters

```python
fig = plt.figure(figsize=(2,2), dpi=120)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

# bbox keeps the axis on the saved image
plt.savefig('new_figure.png',bbox_inches='tight')
```

## Subplots

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
type(axes)
```

```python
axes.shape
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
axes.shape
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes:
        ax.plot(x,y)
# make sure the graphics are not coliding
plt.tight_layout()
```

```python
# all the fig parameters can be used in subplots, like figsize
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(5,5),dpi=120)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)
axes[0][1].set_ylabel('Y LABEL 0,1')

axes[1][0].plot(x,y)
axes[1][0].set_title('TITLE 1,0')

axes[1][1].plot(a,b)

fig.suptitle("Figure Level", fontsize=16)

# Manually set space between graphics
fig.subplots_adjust(wspace=0.9,hspace=0.9)

plt.savefig("new_subplots.png", bbox_inches='tight')
```

## Legends

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc=0);
```

## Styling

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e6c017',lw=10,ls=':')

ax.plot(x,x+1,color='purple',linewidth=2,linestyle='-.')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x+1,color='purple',linewidth=5)
lines[0].set_dashes([1,2,1,2,10,2])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x+1,color='purple',linewidth=2,linestyle='--', marker='o', markersize=20)
```
