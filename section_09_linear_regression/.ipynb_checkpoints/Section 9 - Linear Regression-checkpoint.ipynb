# Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
sns.scatterplot(x='total_spend',y='sales',data=df)
```

```python
# Basically, we want to figure out how to create this line
sns.regplot(x='total_spend',y='sales',data=df)
```

```python
X = df['total_spend']
y = df['sales']
```

```python
help(np.polyfit)
```

```python
# Returns highest order coef first!
np.polyfit(X,y,1)
```

```python
# Potential Future Spend Budgets
potential_spend = np.linspace(0,500,100)
```

```python
predicted_sales =  0.04868788*potential_spend + 4.24302822
```

```python
plt.plot(potential_spend,predicted_sales)
```

```python
sns.scatterplot(x='total_spend',y='sales',data=df)
plt.plot(potential_spend,predicted_sales,color='red')
```

```python
spend = 200
predicted_sales =  0.04868788*spend + 4.24302822
```

```python
predicted_sales
```

```python
np.polyfit(X,y,3)
```

```python
# Potential Future Spend Budgets
potential_spend = np.linspace(0,500,100)
```

```python
predicted_sales =   3.07615033e-07*potential_spend**3 + -1.89392449e-04*potential_spend**2 + 8.20886302e-02*potential_spend**1 + 2.70495053e+00
```

```python
sns.scatterplot(x='total_spend',y='sales',data=df)
plt.plot(potential_spend,predicted_sales,color='red')
```

# Linear Regression with SciKit-Learn

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

```python
sns.pairplot(df)
```

```python
X = df.drop('sales',axis=1)
```

```python
X
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
len(df)
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#help(LinearRegression)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```
