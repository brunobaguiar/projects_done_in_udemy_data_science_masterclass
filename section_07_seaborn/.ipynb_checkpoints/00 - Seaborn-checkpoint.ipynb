# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/dm_office_sales.csv')
```

```python
df.head()
```

## Scatterplots

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education');
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, hue='salary');
```

Use this link to look for color palettes

https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education', 
                palette='Dark2');
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, size='work experience',alpha=0.2);
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, style='level of education',hue='level of education');

plt.savefig('myplot.png')
```

## Distribution Plots

There are 3 main distributions:
* Rug Plot
* Histogram
* KDE Plot

```python
df = pd.read_csv('../DATA/dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
# height is hust visual
```

```python
sns.set(style='darkgrid')
sns.displot(data=df, x='salary',bins=20,color='red',edgecolor='blue',linewidth=4,ls='--',kde=True)
```

```python
sns.histplot(data=df,x='salary')
```

```python
sns.kdeplot(data=df,x='salary',clip=[40000,160000],bw_adjust=0.6,shade=True)
```

## Categorical Plots - Statistics within categories

```python
df = pd.read_csv('../DATA/dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='division')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='level of education',hue='division', palette="Set2")
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd',
           hue='division')

plt.legend(bbox_to_anchor=(1.05,1))
```

## Categorical Plots - Distribution with categories

```python
df = pd.read_csv('../DATA/StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.boxplot(data=df,y='math score')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='math score',x='parental level of education',
           hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(6,6),dpi=200)
sns.boxplot(data=df,x='math score',y='parental level of education',
           hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(6,6),dpi=200)
sns.violinplot(data=df,x='math score',y='parental level of education')
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(data=df,x='math score',y='gender',size=2,
             hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.35,0.5))
```

```python
sns.boxenplot(data=df,x='math score',y='test preparation course',hue='gender')
```

## Comparison Plots

```python
df = pd.read_csv('../DATA/StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.set(style="white")
sns.jointplot(data=df,x='math score',y='reading score',kind='hex')
```

```python
sns.set(style="white")
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',shade=True)
```

```python
sns.set(style="white")
sns.jointplot(data=df,x='math score',y='reading score',hue='gender')
```

```python
sns.pairplot(df,hue='gender',corner=True)
```

## Seaborn Grids

```python
df = pd.read_csv('../DATA/StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',row='lunch',col='test preparation course')
```

```python
g = sns.PairGrid(df, hue='gender')

g= g.map_upper(sns.scatterplot)
g= g.map_lower(sns.kdeplot)
g= g.map_diag(sns.histplot)

g= g.add_legend()
```
